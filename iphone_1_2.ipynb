```python
#2- star rating for some of the models is missing in the dataset.
# fill those missing values with the average rating all the models.

df.loc[:,'star_rating']
df.loc[:,'star_rating'].head(10)
```

```python
# Calculate the average rating excluding None (missing values)
average_rating = df['star_rating'].mean(skipna=True)
```

```python
# Fill missing values with the calculated average rating
df['star_rating'].fillna(average_rating, inplace=True)
```
